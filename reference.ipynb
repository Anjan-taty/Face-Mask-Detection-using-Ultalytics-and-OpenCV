##### Below code is for getting the live video and making a bounding box around face. which is Facial Detection in short

```python
import cv2

# Load Haar Cascade properly (portable path)
face_cap = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml" # type: ignore
)

# Start webcam
video_cap = cv2.VideoCapture(0)

while True:
    ret, frame = video_cap.read()

    if not ret:
        print("Camera not working")
        break

    # Flip image (fix mirror issue)
    frame = cv2.flip(frame, 1)

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces (more strict)
    faces = face_cap.detectMultiScale(
        gray,
        scaleFactor=1.2,     # ↑ better accuracy
        minNeighbors=8,      # ↑ reduce false detection
        minSize=(40, 40)
    )

    # Draw detections
    for (x, y, w, h) in faces:

        # # Rectangle (blue color)
        # cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 255, 0), 3)

        # OR Circle (uncomment if needed)
        center = (x + w//2, y + h//2)
        radius = w // 2
        cv2.circle(frame, center, radius, (0, 0, 255), 3)

    # Show frame
    cv2.imshow("Video_live", frame)

    # Proper exit key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release camera + close window (VERY IMPORTANT)
video_cap.release()
cv2.destroyAllWindows()
```
